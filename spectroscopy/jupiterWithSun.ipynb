```python
# import modules
from specutils import Spectrum1D
from specutils.fitting import fit_generic_continuum


from matplotlib import pyplot as plt
%matplotlib inline

import numpy as np

from astropy.io import fits
from astropy import units as u
from astropy.nddata import StdDevUncertainty

import os
```

```python
infile = "Jupiter_OBJECT_CROP.fit"

jup = Spectrum1D.read(infile,unit=u.adu)
```

```python
# plot spectrum 

plt.figure(figsize=(10,8))
plt.plot(jup.wavelength,jup.flux)
plt.xlabel(r'$\rm Wavelength \ (\AA)$',fontsize=20)
plt.ylabel(r'$\rm Normalized \ Flux$',fontsize=20)
plt.title('LISA Spectrum of Jupiter',fontsize=22)
```
